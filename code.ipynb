```python
def age_distibution_plot(df):
    '''
    Plots the distribution of age groups in the provided DataFrame.

    This function creates a bar chart displaying the count of entries for each age 
	group in the DataFrame, with a horizontal line indicating the mean across age 
	groups.

	Parameters:
    -----------
    df : pandas.DataFrame
        The input DataFrame that contains an 'Age_Group' column. 

	Returns:
    --------
    None
        This function does not return any value. It displays a plot showing the 
		distribution of counts across different age groups.
    '''
    # Define the data
    age_group = df.groupby("Age_Group").size()
    age_group.sort_index(inplace=True)

    # Intialize the figure
    plt.figure(figsize=(10, 6))

    # Call the bar function
    plt.bar(age_group.index, age_group.values)

    # Axis Labelling
    plt.xlabel("Age")
    plt.ylabel("Count")

    # Define mean horizontal line
    plt.axhline(y=np.median(age_group), color = 'r',linestyle = '--', label="mean")
    
    # Display Title
    plt.title('Age Distribution')

    # Display Legend
    plt.legend()

    # Display the plot
    plt.show()
```

```python
def socioeconomic_status_plot(df):
	'''
	Plots the trend of smoking prevalence over time for different socioeconomic 
	statuses.

    This function generates a line plot to illustrate the average smoking 
	prevalence across different years, segmented by socioeconomic status groups. 
	Each line represents a unique socioeconomic status category and shows how 
	smoking prevalence has changed over time.

    Parameters:
    -----------
    df : pandas.DataFrame
        The input DataFrame containing the following columns:
        - 'Socioeconomic_Status': Categorical variable indicating the socioeconomic status.
        - 'Year': The year associated with each entry.
        - 'Smoking_Prevalence': Numeric values representing the smoking prevalence 
			rate.

    Returns:
    --------
    None
        This function does not return any value. It displays a line plot that shows
		the trend of smoking prevalence for each socioeconomic status group across 
		years.
	'''
	
	# Define the data
	year = df.groupby(['Socioeconomic_Status', "Year"])["Smoking_Prevalence"]\
			.mean().reset_index()
	grouped = year.groupby('Socioeconomic_Status')

	# Intialization of the figure
	plt.figure(figsize=(10, 6))

	# Calling the plot function to display lineplot
	for name, group in grouped:
		plt.plot(group["Year"], group["Smoking_Prevalence"], label=name)

	# Axis labelling
	plt.xlabel("Year")
	plt.ylabel("Socioeconomic Status")
	
	# Display Title
	plt.title("Socioeconomic Status with Year")

	# Display legend
	plt.legend()

	# Display the plot
	plt.show()
```

```python
def smoking_drug_plot(df):
	'''
	Plots a boxplot comparing the distributions of smoking prevalence and drug 
	experimentation.

    This function creates a boxplot to compare the distributions of the 
    'Smoking_Prevalence' and 'Drug_Experimentation' columns in the provided 
	DataFrame.

    Parameters:
    -----------
    df : pandas.DataFrame
        The input DataFrame that contains the following columns:
        - 'Smoking_Prevalence': Numeric values representing smoking prevalence 
			rates.
        - 'Drug_Experimentation': Numeric values indicating drug experimentation 
			levels.

    Returns:
    --------
    None
        This function does not return any value. It displays a boxplot showing the 
		distribution of values for smoking prevalence and drug experimentation.
	'''
	
	# Caling the boxplot
	df.boxplot(["Smoking_Prevalence", "Drug_Experimentation"])

	# Display Title
	plt.title('Smoking vs Drug')

	# Axis labelling
	plt.ylabel("Value")

	# Display the plot
	plt.show()
```

```python
# Calling the function
socioeconomic_status_plot(df)
```

```python
# Calling the function
smoking_drug_plot(df)
```
